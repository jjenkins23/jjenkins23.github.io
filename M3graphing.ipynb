```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MultipleLocator

df_c = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
```

```python
fig,ax = plt.subplots()

ax.plot(df_c['Month-Year'], df_c['Total Construction'], color="k")
ax.plot(df_c['Month-Year'], df_c['Private Construction'], color="b")
ax.plot(df_c['Month-Year'], df_c['Public Construction'], color="r")

plt.xticks(range(0, 146, 12), rotation='vertical', fontsize=14)
ax.xaxis.set_minor_locator(MultipleLocator(6))
ax.grid(which="both")
ax.set_xlim(0,146) 
fig.suptitle('Construction Time Series', weight='bold', fontsize=24)   # Graph title
ax.xaxis.set_label_text('Month-Year', weight='bold', fontsize=18)  # x-axis caption
ax.yaxis.set_label_text('Construction (Thousands of $)', weight='bold', fontsize=18)       # y-axis caption
ax.set_yticklabels(["${:,.2f}".format(y) for y in ax.get_yticks()], fontsize=14)
fig.set_size_inches(10,6)
ax.legend(['Total', 'Private', 'Public'],  fontsize=16)
plt.show()
# start, end = ax.get_xlim()
#print(start, end)
```

### M3 Advanced Graphing
The graph is shown 3 time monthly series over 12 year period for public and private construction as well as their total.
The important aspect of the time series is its cyclic trend over year.
Major axis tick marks and grid lines directly hit the lower points of the graph for every year in January, and the minor ticks in between correspond to the higher points in the middle of each year (July). Also seems like there is long cycle in private construction with the cycle length about 10 years. Private constructions declined after 2008 crisis and began recovering in 2011-2014.

**Matplotlib Features used:**
 - Custom Series colorinng
 - Legend
 - Custom font size
 - Custom tick mark location and frequency
 - Custom tick mark format and rotation
 - Grid lines over major and minor ticks
 - Axes labels and plot title
